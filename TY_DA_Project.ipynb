```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Now we will plot number of School vs States in India

```python
plt.figure(figsize=(10,12))
sns.barplot(df['SCHTOT'], df['STATNAME'],alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Number of Schools', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Number of schools wrt states in India", fontsize=16)
plt.show()
```

#From the last 2 graphs we can analyze that when population increases the number of schools also increases

# **We will now see the ratio of government school vs Private School plotting in a pie plot**

```python
plt.figure(figsize=(20,12))
for i in range(1,len(df)):
    plt.subplot(4,9,i)
    plt.title(df['STATNAME'][i])
    top = ['Gov','pri']
    uttar = df.loc[df['STATNAME'] == df['STATNAME'][i],:]
    value =[float(uttar['SCHTOTG']/uttar['SCHTOT'])*100,float(uttar['SCHTOTPR']/uttar['SCHTOT'])*100]
    plt.pie(value, labels=top, autopct='%1.1f%%',startangle=140)
    plt.axis('equal')
plt.show()
```

##We can analze that the government of india has done quite an impressive work uptil now since the number of government schools are exceeding the number of private schools upto a great extend

Finally we have plotted the literacy rate per state

```python

plt.figure(figsize=(10,12))
sns.barplot( df['OVERALL_LI'],df['STATNAME'],
order=df.sort_values('OVERALL_LI',ascending = False).STATNAME)
plt.xticks(rotation='vertical')
plt.xlabel('Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Literacy Rate wrt states in India", fontsize=16)
plt.show()
```

#From the last graph of literacy per state we can see that states of kerala, lakshadweep, mizoram, tripura, goa, daman and diu, puducherry, chandigarh and delhi have highest literacy rates.
#States of Bihar, Telangana, Arunachal Pradesh, Rajasthan, Jharkhand, Andhra Pradesh, Jammu Kashmir, uttar pradesh and madhya pradesh have quite a less literacy rates and need to focus more in education field.

```python
df['good'] = df['ROADTOT'] +df['SPLAYTOT'] +df['SWATTOT'] +df['SELETOT']
df['goodpercent'] = df['good']/df['SCHTOT']
plt.figure(figsize=(10,12))
sns.barplot(df['goodpercent'], df['STATNAME'],alpha=1)
plt.xticks(rotation='vertical')
plt.xlabel("Literacy rate", fontsize=16)
plt.title('Literacy rate with respect to state')
plt.show()
```

Percentage of schools with Playground Facility and Drinking Water and Electricity and also approachable by All Weather Roads in almost all states in India is < 4% which is pretty low

The percentage of higly equipped schools are highest in the state of punjab and lowest in the state of meghalaya

```python
top_5=df.sort_values(by='OVERALL_LI', ascending=False).head(5)
top_5
```

```python
bottom_5=df.sort_values(by='OVERALL_LI', ascending=True).head(5)
bottom_5
```

```python
top_bottom=pd.concat([top_5, bottom_5], axis=0, sort=False)
top_bottom
```

```python
top_bottom['Pop_DEN']=(top_bottom['TOTPOPULAT']/top_bottom['AREA_SQKM'])*1000

curr_cols=pd.concat([top_bottom['STATNAME'],top_bottom['Pop_DEN']],axis=1,sort=False)
curr_cols
```

```python
top_bottom['Diff_Lit']=top_bottom['MALE_LIT']-top_bottom['FEMALE_LIT']
curr_cols=pd.concat([top_bottom['STATNAME'],top_bottom['Diff_Lit']],axis=1,sort=False)
curr_cols
```

Let's plot this data into bar graph:

```python
plt.figure(figsize=(16,7))
sns.barplot( top_bottom['STATNAME'],top_bottom['Diff_Lit'], alpha=0.9)
plt.title("Lteracy rate f Vs M")
```

```python
top_bottom['P_RUR_POP']=100-top_bottom['P_URB_POP']
curr_cols=pd.concat([top_bottom['STATNAME'],top_bottom['P_RUR_POP']],axis=1,sort=False)
curr_cols
```
